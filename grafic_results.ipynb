```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

# Pruebas

## Servidor Real Sistemas

```python
# Leer el archivo CSV
df_real = pd.read_csv('./results/real_server_results.jtl', sep=',', skip_blank_lines=True, on_bad_lines='warn')

# Filtrar las filas cuyo responseCode no empiece por 3
df_real = df_real[~df_real['responseCode'].astype(str).str.startswith('3')]
```

```python
# Convertir el timestamp a un formato de fecha y hora
df_real['timeStamp'] = pd.to_datetime(df_real['timeStamp'], unit='ms')

# Filtrar los datos por éxito y fallo
success_df_all_all_all_all_all_all_all_all = df_real[df_real['success'] == True]
failure_df = df_real[df_real['success'] == False]

# Crear la figura y los ejes
plt.figure(figsize=(6, 5))

# Graficar los tiempos de respuesta exitosos
plt.plot(success_df['timeStamp'], success_df['elapsed'], 'g.', label='Aciertos (ms)')

# Graficar los tiempos de respuesta fallidos
plt.plot(failure_df['timeStamp'], failure_df['elapsed'], 'r.', label='Errores (ms)')

# Configurar etiquetas y título
plt.xlabel('Tiempo')
plt.ylabel('Tiempo de Respuesta (ms)')
plt.title('Aciertos y Errores a lo largo del Tiempo')
plt.legend()
plt.grid(True)

# Formatear el eje x para cortar los decimales del timestamp
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

# Mostrar la gráfica
plt.show()
```

```python
# Contar el número total de filas
total_requests = len(df_real)

# Filtrar las filas donde la columna 'success' es False
errores = df_real[df_real['success'] == False]

# Contar el número de filas con errores
num_errores = len(errores)

# Calcular el porcentaje de errores
porcentaje_error = (num_errores / total_requests) * 100

# Imprimir el porcentaje de errores
print(f'Número total de solicitudes: {total_requests}')
print(f'Porcentaje de error: {porcentaje_error:.2f}%')
```
