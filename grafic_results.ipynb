```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

# Pruebas servidor real de sistemas

## Pruebas de carga con 100 hilos

```python
# Leer el archivo CSV
df_real = pd.read_csv('./results/real_server_results_100.jtl', sep=',', skip_blank_lines=True, on_bad_lines='warn')

# Filtrar las filas donde 'label' tenga un valor específico
df_real = df_real[df_real['label'] == 'sistemas']
```

```python
# Convertir el timestamp a un formato de fecha y hora
df_real['timeStamp'] = pd.to_datetime(df_real['timeStamp'], unit='ms')

# Filtrar los datos por éxito y fallo
success_df = df_real[df_real['success'] == True]
failure_df = df_real[df_real['success'] == False]

# Contar el número total de filas
total_requests = len(df_real)
num_errores = len(failure_df)

# Calcular el porcentaje de errores
porcentaje_error = (num_errores / total_requests) * 100

# Calcular el tiempo promedio de respuesta para las solicitudes exitosas
tiempo_promedio_respuesta_exitosas = success_df['elapsed'].mean()

# Imprimir datos
print(f'Número total de solicitudes: {total_requests}')
print(f'Porcentaje de error: {porcentaje_error:.2f}%')
print(f'Tiempo promedio de respuesta: {tiempo_promedio_respuesta_exitosas:.2f} ms')

# Crear la figura
plt.figure(figsize=(6, 5))
plt.plot(success_df['timeStamp'], success_df['elapsed'], 'g.', label='Aciertos (ms)')
plt.plot(failure_df['timeStamp'], failure_df['elapsed'], 'r.', label='Errores (ms)')
plt.xlabel('Tiempo')
plt.ylabel('Tiempo de Respuesta (ms)')
plt.title('Aciertos y Errores a lo largo del Tiempo')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.show()
```
