```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import StandardScaler
```

# Pruebas servidor real de sistemas

## Pruebas de carga con 100 hilos

```python
# Leer el archivo CSV
df_real = pd.read_csv('./results/real_server_results_100.jtl', sep=',', skip_blank_lines=True, on_bad_lines='warn')

# Filtrar las filas donde 'label' tenga un valor específico
df_real = df_real[df_real['label'] == 'sistemas']
```

```python
# Convertir el timestamp a un formato de fecha y hora
df_real['timeStamp'] = pd.to_datetime(df_real['timeStamp'], unit='ms')

# Filtrar los datos por éxito y fallo
success_df = df_real[df_real['success'] == True]
failure_df = df_real[df_real['success'] == False]

# Contar el número total de filas
total_requests = len(df_real)
num_errores = len(failure_df)

# Calcular el porcentaje de errores
porcentaje_error = (num_errores / total_requests) * 100

# Calcular el tiempo promedio de respuesta para las solicitudes exitosas
tiempo_promedio_respuesta_exitosas = success_df['elapsed'].mean()

# Imprimir datos
print(f'Número total de solicitudes: {total_requests}')
print(f'Porcentaje de error: {porcentaje_error:.2f}%')
print(f'Tiempo promedio de respuesta: {tiempo_promedio_respuesta_exitosas:.2f} ms')

# Crear la figura
plt.figure(figsize=(6, 5))
plt.plot(success_df['timeStamp'], success_df['elapsed'], 'g.', label='Aciertos (ms)')
plt.plot(failure_df['timeStamp'], failure_df['elapsed'], 'r.', label='Errores (ms)')
plt.xlabel('Tiempo')
plt.ylabel('Tiempo de Respuesta (ms)')
plt.title('Aciertos y Errores a lo largo del Tiempo')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.show()
```

## Comparación

```python
# Calcular el porcentaje de aciertos y errores para cada caso
def calculate_percentages(df):
    total = len(df)
    success = len(df[df['success'] == True])
    error = len(df[df['success'] == False])
    success_percentage = (success / total) * 100
    error_percentage = (error / total) * 100
    return success_percentage, error_percentage

# Calcular porcentajes para cada caso
success_new_100, error_new_100 = calculate_percentages(df_new_100)
success_new_500, error_new_500 = calculate_percentages(df_new_500)
success_new_1000, error_new_1000 = calculate_percentages(df_new_1000)

# Crear un DataFrame con los porcentajes
data = {
    'Caso': ['Contenedor 100 hilos', 'Contenedor 500 hilos', 'Contenedor 1000 hilos'],
    'Aciertos': [success_new_100, success_new_500, success_new_1000],
    'Errores': [error_new_100, error_new_500, error_new_1000]
}
df_percentages = pd.DataFrame(data)

# Crear el diagrama de barras
plt.figure(figsize=(7, 5))
df_percentages_melted = df_percentages.melt(id_vars='Caso', value_vars=['Aciertos', 'Errores'], var_name='Resultado', value_name='Porcentaje')
sns.barplot(x='Caso', y='Porcentaje', hue='Resultado', data=df_percentages_melted)
plt.title('Porcentaje de Aciertos y Errores para Diferentes Casos')
plt.ylabel('Porcentaje')
plt.xlabel('Caso')
plt.legend(title='Resultado')
plt.grid(True)
plt.show()
```

```python
# Filtrar los datos por éxito y error
success_df_real = df_real[df_real['success'] == True]
success_df_new_100 = df_new_100[df_new_100['success'] == True]

scaler = StandardScaler()
success_df_real['elapsed'] = scaler.fit_transform(success_df_real[['elapsed']])
success_df_new_100['elapsed'] = scaler.fit_transform(success_df_new_100[['elapsed']])

# Crear subplots
fig, axes = plt.subplots(1, 2, figsize=(6, 5))

# Diagrama de caja para tiempos de respuesta de éxitos
sns.boxplot(y='elapsed', data=success_df_real, ax=axes[0])
axes[0].set_ylabel('Tiempo de Respuesta (ms)')
axes[0].set_xlabel('Servidor real')
axes[0].grid(True)

# Diagrama de caja para tiempos de respuesta de errores
sns.boxplot(y='elapsed', data=success_df_new_100, ax=axes[1])
axes[1].set_ylabel('Tiempo de Respuesta (ms)')
axes[1].set_xlabel('Contenedor')
axes[1].grid(True)

# Ajustar el layout
fig.suptitle('Comparación de Tiempos de Respuesta entre\nServidor Real y Contenedor con 100 hilos', fontsize=12)
plt.tight_layout(rect=[0, 0, 1, 0.99])
plt.show()
```
